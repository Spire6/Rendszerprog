```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

# Required to use matplotlib embedded in Jupyter notebook
%matplotlib inline

# set inline graph size
plt.rcParams['figure.figsize'] = [15, 5]
```

## Creating 2 clusters of random data 

```python
# cluster centers
c1 = [6, 2]
c2 = [2, 6]

# create point cloud in the Origo
X = np.zeros((100, 2))

# create random labels
y = np.random.rand(1, 100).ravel() > 0.5

# move positive labeled points to near 1st cluster center 
X[y] = X[y] + c1

# move negative labeled points to near 2st cluster center 
X[np.invert(y)] = X[np.invert(y)] + c2
```

## Put random noise on point cloud

```python
X_rnd = np.random.normal(loc=0, scale=1.0, size=(100, 2))
X = X + X_rnd
```

## Visualize labeled data

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.show()
```

```python
k_means = KMeans(n_clusters=2)
k_means.fit(X)

print(k_means.n_iter_)

plt.scatter(X[k_means.labels_ == 0][:, 0], X[k_means.labels_ == 0][:, 1], s=50, c='b', marker='s')
plt.scatter(X[k_means.labels_ == 1][:, 0], X[k_means.labels_ == 1][:, 1], s=50, c='g', marker='^')
plt.scatter(k_means.cluster_centers_[0, 0], k_means.cluster_centers_[0, 1], s=200, c='y', marker='o')
plt.scatter(k_means.cluster_centers_[1, 0], k_means.cluster_centers_[1, 1], s=200, c='r', marker='o')
plt.show()
```
